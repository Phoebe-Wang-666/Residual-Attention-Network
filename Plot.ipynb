```python
import pandas as pd

df=pd.read_excel("one-Validation Top-1 Accuracy .xlsx")
df.head()
```

ARL

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10,6))

plt.plot(df["Epoch"],df["attention56_arl"],label="56")
plt.plot(df["Epoch"],df["attention92_ar"],label="92")
plt.plot(df["Epoch"],df["attention128_arl"],label="128")
plt.plot(df["Epoch"],df["attention164_arl"],label="164")

plt.xlabel("epoch")
plt.ylabel("val_accuracy")
plt.title("ARL")
plt.legend(title="num_attention_layer")
plt.grid(True)

plt.tight_layout()
plt.show()
```

NAL

```python
plt.figure(figsize=(10,6))

plt.plot(df["Epoch"],df["attention56_nal"],label="56")
plt.plot(df["Epoch"],df["attention92_nal"],label="92")
plt.plot(df["Epoch"],df["attention128_nal"],label="128")
plt.plot(df["Epoch"],df["attention164_nal"],label="164")

plt.xlabel("epoch")
plt.ylabel("val_accuracy")
plt.title("NAL")
plt.legend(title="num_attention_layer")
plt.grid(True)

plt.tight_layout()
plt.show()
```

Compare ARL vd NAL

```python

plt.figure(figsize=(10,6))

plt.plot(df["Epoch"],df["attention56_arl"],label="56_arl")
plt.plot(df["Epoch"],df["attention56_nal"],label="56_nal")


plt.xlabel("epoch")
plt.ylabel("val_accuracy")
plt.title("Attention 56: ARL vs NAL")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
